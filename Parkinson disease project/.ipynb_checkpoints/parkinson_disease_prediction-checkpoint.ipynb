```python
## Importing dependencies
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm 
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# 5. Selecting features and target variable
X = parkinson_disease.drop(columns=['name', 'status'])  # Drop 'name' and 'status' as they are not needed
y = parkinson_disease['status']  # 'status' is the target variable
```

```python
parkinson_disease.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Function to evaluate models
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    print(f"\n{model_name}")
    print(f"Accuracy: {acc:.4f}")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Classification Report:\n{report}")
    plot_confusion_matrix(cm, model_name)
    return model
```

```python
# Function to plot confusion matrix
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - {title}')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
# SVM
svm_model = svm.SVC(kernel='linear')
evaluate_model(svm_model, X_train, X_test, y_train, y_test, "Support Vector Machine (SVM)")
```

```python
# Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
evaluate_model(dt_model, X_train, X_test, y_train, y_test, "Decision Tree Classifier")
```

```python
# Logistic Regression
lr_model = LogisticRegression(random_state=42)
evaluate_model(lr_model, X_train, X_test, y_train, y_test, "Logistic Regression")
```

```python
# Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
evaluate_model(gb_model, X_train, X_test, y_train, y_test, "Gradient Boosting Classifier")
```

```python
# Logistic Regression with increased max_iter and different solver
print("\nLogistic Regression:")
lr_model = LogisticRegression(random_state=42, max_iter=1000, solver='saga')
evaluate_model(lr_model, X_train, X_test, y_train, y_test, "Logistic Regression")

# Cross-validation scores
print("\nCross-validation scores:")
models = {
    'SVM': svm_model, 
    'Decision Tree': dt_model, 
    'Logistic Regression': lr_model, 
    'Gradient Boosting': gb_model
}
cv_scores = []
for model_name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5)
    mean_score = np.mean(scores)
    cv_scores.append((model_name, mean_score))
    print(f"{model_name} Cross-validation Accuracy: {mean_score:.4f}")
```

```python
# Visualizing cross-validation scores with percentages
model_names, scores = zip(*cv_scores)

# Convert accuracy scores to percentages
percentages = [score * 100 for score in scores]

# Create the bar plot
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=percentages, y=model_names, palette='coolwarm')

# Add percentage labels on top of the bars
for i, v in enumerate(percentages):
    ax.text(v + 1, i, f'{v:.2f}%', color='black', ha="left", va="center")

plt.xlabel('Cross-validation Accuracy (%)')
plt.title('Cross-validation Accuracy Scores for Different Models')
plt.show()
```

```python
# List of models
models = {
    'SVM': svm_model, 
    'Decision Tree': dt_model, 
    'Logistic Regression': lr_model, 
    'Gradient Boosting': gb_model
}

# Calculate accuracy for each model and store as percentage
model_accuracies = {name: accuracy_score(y_test, model.fit(X_train, y_train).predict(X_test)) * 100 
                    for name, model in models.items()}

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=list(model_accuracies.values()), y=list(model_accuracies.keys()), palette='coolwarm')

# Adding percentage labels
for i, v in enumerate(model_accuracies.values()):
    plt.text(v + 1, i, f'{v:.2f}%', color='black', va="center")

plt.xlabel('Accuracy (%)')
plt.title('Model Accuracy Comparison (Hold-Out Method)')
plt.show()
```
